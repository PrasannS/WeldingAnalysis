```python
from bs4 import BeautifulSoup
import requests
```

```python
page_link = 'http://www.ceweld.nl/en/catalogue?field_product_cat_tid=All&title=&field_werkstofnummer_value=&field_aws_value=&field_aws_value_1=All&field_din_value=&field_iso_value=&field_iso_value_1=All&field_material_value=&items_per_page=All'

page_response = requests.get(page_link, timeout=5)

page_content = BeautifulSoup(page_response.content, "html.parser").body

textContent = []
```

```python
textContent= []
for i in range (36, 10000):
    z = page_content.find_all("a", href=True)
    print(z[i]['href'])
    textContent.append(z[i]['href'])
```

```python
print(textContent)
arr = []
for s in textContent:
    if s.startswith('/'):
        arr.append(s)
        
print(arr)
```

```python
#'http://www.ceweld.nl/en/product/cusi-3-laser-brazing-welding-wire'
def get_data(link):
    page_link = link
    page_response = requests.get(page_link, timeout=5)
    soup = BeautifulSoup(page_response.content, "html.parser").body
    textContent = []
    a = {}
    a = {
        'Al': None,
        'Cu': None,
        'Mn': None,
        'Si': None,
        'Zr': None,
        'Ti': None,
        'V': None,
        'Zn': None,
        'Si': None,
        'C': None,
        'Cr': None,
        'Ni': None,
        'Mo': None,
        'P': None,
        'S': None,
        'Nb': None,
        'Fe': None,
        'Other':None
    }
    i=0
    table = soup.find_all("table", id='tablefield-0')[0]
    for th in table.find_all("th"):
        a[th.find_all("p")[0].text]=table.find_all("td")[i]
        i++
    
    return a;
```

```python
for i in arr:
    t=[]
    print(i)
    t = get_data('http://www.ceweld.nl'+i)
    print(t)
    data = data.append(t, ignore_index=True)
    
print(data)
```
